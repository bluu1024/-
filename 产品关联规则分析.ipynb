# 二、数据预处理

## 2.1构造购物篮事务库

```python
#汇总同一日期同一客户ID的消费记录
col=['订单日期','客户ID','区域','产品名称']

def gather(x):  # 创建函数将x返回列表
    return list(x)
df=order.loc[:,col].groupby(['订单日期','客户ID','区域']).agg(gather).reset_index() # 分组聚合后设置索引
df
```

```python
#将同一消费记录中的商品名称唯一化
df_s=df.copy()
df_s['产品名称']=df['产品名称'].apply(lambda x: list(set(x)))
df_s
```

```python
#构建购物篮事务库
transactions=[]
for i in df_s.index:
    transactions.append(list(df_s.产品名称[i])) #将每一条消费记录中的产品名称转换为列表
transactions=list(transactions)
transactions
```

## 2.2将数据类型转换为布尔值构成的dataframe

```python
from mlxtend.preprocessing import TransactionEncoder
TE=TransactionEncoder() #构造转换模型
df_s1=TE.fit_transform(transactions) #将原始数据转换为布尔值
df_s1=pd.DataFrame(df_s1,columns=TE.columns_)
df_s1
```

# 三、关联规则Apriori算法应用

## 3.1 按总体销售计算频繁项集挖掘强关联规则

```python
#利用Aprior算法计算频繁项集（最小支持度min_support自定义）
from mlxtend.frequent_patterns import apriori
items=apriori(df_s1,min_support=0.02,use_colnames=(True))
items
```

```python
#按支持度排序（倒序）查看频繁项集
items.sort_values(by='support',ascending=False).head()
```

```python
#根据最小置信度在频繁项集中挖掘强关联规则(最小置信度min_threshold)
from mlxtend.frequent_patterns import association_rules
rules=association_rules(items,min_threshold=0.3)
rules
```

## 3.2 按销售区域分区计算频繁项集挖掘强关联规则

```python
# 重复数据预处理阶段 
''' 
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

col=['订单日期','客户ID','区域','产品名称']
def gather(x):
    return list(x + '')
df=order.loc[:,col].groupby(['订单日期','客户ID','区域']).agg(gather).reset_index()

#将同一消费记录中的商品名称唯一化
df_s=df['产品名称']=df['产品名称'].apply(lambda x: list(set(x)))
df_s
'''
area=['大中华区','韩国','新加坡']

for q in area:
    df_s=df.loc[df['区域']==q]
    
    # 构建购物篮事务库
    transactions=[]
    for i in df_s.index:
        transactions.append(list(df_s.产品名称[i])) 
    transactions=list(transactions)
    
    # 将数据类型转换为布尔值构成的dataframe
    TE=TransactionEncoder() #构造转换模型
    df_s1=TE.fit_transform(transactions) #将原始数据转换为布尔值
    df_s1=pd.DataFrame(df_s1,columns=TE.columns_)
    
    # 利用Aprior算法计算频繁项集（最小支持度min_support=0.03）
    items=apriori(df_s1,min_support=0.03,use_colnames=(True))
    print(q+ '区域：')
    print('频繁项集：')
    print(items)  
```

```python
# 按支持度排序（倒序）查看频繁项集
items.sort_values(by='support',ascending=False).head(10) 
```

```python
#根据最小置信度在频繁项集中挖掘强关联规则(最小置信度min_threshold=0.3）
from mlxtend.frequent_patterns import association_rules
rules=association_rules(items,min_threshold=0.3)
rules
```

# 商品分析

## 4.1 商品区域销售情况分析

```python
# 透视各区域、各国家、各个销售大区的总销量、总利润等情况
order.pivot_table(values=['订单数量','产品成本','利润','销售金额'],index=['区域','国家','销售大区'],aggfunc(np.sum))
```

```python
#统计各个区域的总销量、总利润等占比情况（3个区）
order_qy=order.pivot_table(values=['订单数量','产品成本','利润','销售金额'],index=['区域'],aggfunc=(np.sum))

order_qy_p=order_qy.div(order_qy.sum(axis=0),axis=1).sort_values(by='订单数量',ascending=False)
order_qy_p

#绘制饼图
fig=plt.figure(figsize=(15,15))
i=1  
for c in order_qy_p.columns:
    ax=fig.add_subplot(2,2,i)
    plt.pie(order_qy_p[c],labels=order_qy_p.index,autopct='%1.1f%%')
    plt.title('各区域' + c + '占比')
    i=i+1
plt.show()
```

```python
#统计各个国家的总销量、总利润等占比情况
order_gj=order.pivot_table(values=['订单数量','产品成本','利润','销售金额'],index=['国家'],aggfunc=(np.sum))

order_gj_p=order_gj.div(order_gj.sum(axis=0),axis=1).sort_values(by='订单数量',ascending=False)
order_gj_p
```

## 4.3 商品盈利能力分析
